# PBH Merger Rate

```python
def localrate_Huang(pbhmass,fpbh):
    '''
    unit: Gpc^-3 year^-1
    '''
    sigma_eq = 0.005
    return 3e6*pbhmass**(-32./37)*fpbh**2*(0.7*fpbh**2+sigma_eq**2)**(-21./74)

def allmasslocalrate_Huang(m1,m2,fpbh,p1):
    '''
    Note that here fpbh is the sum of fraction of two masses PBHs
    '''
    sigma_eq = 0.005
    p2 = 1-p1
    return 3e6*fpbh**2*(0.7*fpbh**2+sigma_eq**2)**(-21./74)*min(p1/m1,p2/m2)*(p1/m1+p2/m2)*(m1*m2)**(3./37)*(m1+m2)**(36./37)
```

# Constraint on $f_\mathrm{PBH}$

```python
from scipy import optimize

ligo = np.loadtxt('./data/subsolarligo.txt')
nonecc = np.loadtxt('./data/nonecc.txt')

result = []
pbhmass = con.mass1_from_mchirp_q(nonecc[:,0],1)
```

# Reproduce LIGO results and make our results for equal mass

```python
result = []
pbhmass = con.mass1_from_mchirp_q(nonecc[:,0],1)

for i,_ in enumerate(pbhmass):
    def CacheFunction(f):
        return localrate_Huang(pbhmass[i],f) - nonecc[i,1]
    sol = optimize.root_scalar(CacheFunction, bracket=[1e-4, 1], method='brentq')
    result.append(sol.root)

result_ligo = []
ligomass = ligo[:,0]

for i,_ in enumerate(ligomass):
    def CacheFunction(f):
        return localrate_Huang(ligomass[i],f) - ligo[i,2]
    sol = optimize.root_scalar(CacheFunction, bracket=[1e-4, 1], method='brentq')
    result_ligo.append(sol.root)
```

# Constraint for 7 $M_\odot$

```python
def CacheFunction(f):
    return localrate_Huang(7,f) - 12
sol = optimize.root_scalar(CacheFunction, bracket=[1e-4, 1], method='brentq')
fpbh7 = sol.root
```

```python
nonecc_interp = np.loadtxt('./data/nonecc-interp.txt')

result7 = []
m2 = con.mass2_from_mchirp_mass1(nonecc_interp[:,0],7)

for i,_ in enumerate(m2):
    def CacheFunction(p1):
        return allmasslocalrate_Huang(7,m2[i],fpbh7/p1,p1) - nonecc_interp[i,1]
    sol = optimize.root_scalar(CacheFunction, bracket=[1e-8, 1], method='brentq')
    result7.append(fpbh7/sol.root*(1-sol.root))
```
